```python
import diffusers 
from diffusers import AutoPipelineForText2Image
import torch
pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sd-turbo", variant = "fp16")
```

```python
prompt = "A photo of a deer of in a simple outdoor setting"
```

```python
from PIL import Image
import os

output_folder = "/code/COE379L-Final/DeerAI"
os.makedirs(output_folder, exist_ok=True)

for i in range(3000):
    # Generate the image`
    image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]

    # Save as JPEG
    output_path = f"{output_folder}/deer_{i}.jpg"
    image.save(output_path)
```
