```python
import diffusers 
from diffusers import AutoPipelineForText2Image
import torch
pipe = AutoPipelineForText2Image.from_pretrained("stabilityai/sd-turbo", variant = "fp16")
```

```python
prompt = "A cat of any species in a simple indoor or outdoor setting similar to CIFAR-10 style"
```

```python
from PIL import Image
import os

output_folder = "/code/COE379L-Final/CatAI"
os.makedirs(output_folder, exist_ok=True)

for i in range(3000):
    # Generate the image`
    image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=0.0).images[0]

    # Convert to RGB (needed for JPEG) 
    image = image.convert("RGB")

    # Save as JPEG
    output_path = f"{output_folder}/cat_{i}.jpg"
    image.save(output_path, quality = 85)
```
